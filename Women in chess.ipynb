# Women in chess

by **Larisa Shapiro**

## Goals:

* Calculate the number of women playing chess on various levels in all world regions for the last 10 years.
* Evaluate the dynamics and predict the numbers for the future.

## Chess regions:

* Africa
* Asia
* Australia and Oceania
* Central America
* China
* Europe
* India
* Northern America
* Russia
* South America


## Chess levels:

* **grandmaster**, 	2300 < *rating*
* **master**, 		2100 < *rating* < 2300
* **strong amateur**,	1900 < *rating* < 2100

*Note*: Ratings less than 1900 are not accurate in many regions.

## Data origins:

International chess federation FIDE.

### Data format:

csv file with the ratings of all female chess players for all the countries for the last 10 years.

### Data preprocessing:

* We split chess regions into countries and create a correspondence: *country - region*.
* We add to the original data the '*region*' column and save the relevant data in .csv format.

### Data by levels:

For each *chess level* and for each *year* we calculate the number of *women* in this level. For each *region* and overall.

## Prediction model:

* We predict the number of women chessplayers for each level.
* We use the **polynomial prediction** model of degree 2 based on 10 points.

### Formula Polynomial Regression:

\begin{equation}
Y_i = \beta_0 + \beta_1 X_i + \beta_2 X_i^2 + \beta_3 X_i^3 + ... + \beta_n X_i^n + \epsilon_i
\end{equation}

## Conclusions:

* The number of women **grandmasters** and **masters** is growing, albeit slowly.
* At the same time the number of **strong amateur** women chessplayers is growing a bit faster.
